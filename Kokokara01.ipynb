```python
import requests 
import json
import csv
```

```python
url = "https://u10sme-api.smrj.go.jp/v1/areas.json"
response = requests.get(url)
jsonData = response.json()
areas = []#エリアコード、都道府県コード、都道府県名
for jsonObj in jsonData["areas"]:
    #print(jsonObj['id'])#areaコード
    for jsonObp in jsonObj['prefectures']:#都道府県コード
        #print(jsonObp['code'] , jsonObp['name'])
        areas.append([jsonObj['id'], jsonObp['code'], jsonObp['name']])
print(areas)
with open('./areas.csv', 'w', encoding="utf_8_sig") as f:
    writer = csv.writer(f)
    writer.writerows(areas)
```

```python
url = "https://u10sme-api.smrj.go.jp/v1/serviceCategories.json"
response = requests.get(url)
jsonData = response.json()
serviceCategories = []#
for jsonObj in jsonData["serviceCategories"]:#営業品目区分
    for jsonObp in jsonObj['serviceKinds']:#営業品目種別
        for jsonObc in jsonObp['services']:#営業品目
            try:
                print(jsonObc['example'])
                serviceCategories.append([jsonObj['id'], jsonObj['name'], jsonObp['id'], jsonObp['name'], jsonObc['id'], jsonObc['name'], jsonObc['code'], jsonObc['example']])
            except :
                serviceCategories.append([jsonObj['id'], jsonObj['name'], jsonObp['id'], jsonObp['name'], jsonObc['id'], jsonObc['name'], jsonObc['code'], ''])
print(serviceCategories)
with open('./serviceCategories.csv', 'w', encoding="utf_8_sig") as f:
    writer = csv.writer(f)
    writer.writerows(serviceCategories)
```

```python
url = "https://u10sme-api.smrj.go.jp/v1/tags.json"
response = requests.get(url)
jsonData = response.json()
tags = []#
for jsonObj in jsonData["tags"]:
    tags.append([jsonObj['id'], jsonObj['name']])
print(tags)
with open('./tags.csv', 'w', encoding="utf_8_sig") as f:
    writer = csv.writer(f)
    writer.writerows(tags)
```

```python
url = "https://u10sme-api.smrj.go.jp/v1/sortOrders.json"
response = requests.get(url)
jsonData = response.json()
sortOrders = []#
for jsonObj in jsonData["sortOrders"]:
    sortOrders.append([jsonObj['id'], jsonObj['name']])
print(sortOrders)
with open('./sortOrders.csv', 'w', encoding="utf_8_sig") as f:
    writer = csv.writer(f)
    writer.writerows(sortOrders)
```
